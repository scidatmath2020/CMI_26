# Reglamento Interno  
## Clínica de Metodología de la Investigación

---

## CAPÍTULO I  
### Disposiciones Generales

#### Artículo 1. Naturaleza del espacio
La **Clínica de Metodología de la Investigación** es un espacio académico digital permanente de acompañamiento metodológico, análisis crítico y formación colaborativa, orientado al desarrollo de investigaciones en distintas etapas.

#### Artículo 2. Carácter del acompañamiento
La Clínica **no sustituye la dirección formal de tesis** ni garantiza la aprobación institucional de proyectos.  
Su función es **formativa, orientadora y crítica**.

#### Artículo 3. Modalidad de participación
La participación es **voluntaria, flexible y no sujeta a ciclos cerrados**.  
Los integrantes pueden incorporarse o retirarse en cualquier momento.

---

## CAPÍTULO II  
### De los participantes

#### Artículo 4. Perfil
Podrán participar personas con proyectos de investigación académica o aplicada, independientemente de su **nivel educativo o disciplina**.

#### Artículo 5. Responsabilidad individual
Cada participante es único responsable de:

- Su proyecto.
- Sus decisiones metodológicas.
- El uso que haga de la retroalimentación recibida.

---

## CAPÍTULO III  
### Derechos de autor y confidencialidad

#### Artículo 6. Autoría intelectual
Todo documento, idea, avance, esquema, dato o argumento compartido es **propiedad intelectual exclusiva** de quien lo presenta.

#### Artículo 7. Prohibiciones expresas
Queda estrictamente prohibido:

- Copiar, reproducir o distribuir trabajos ajenos.
- Usar ideas o textos compartidos sin autorización expresa.
- Grabar, capturar pantalla o almacenar contenidos sin permiso.

#### Artículo 8. Confidencialidad
Toda la información compartida dentro de la Clínica se considera **confidencial**, aun después de abandonar el espacio.

---

## CAPÍTULO IV  
### Mecánica de trabajo

#### Artículo 9. Presentación de avances
Los participantes podrán presentar **avances escritos u orales** conforme a los lineamientos establecidos por la coordinación.

#### Artículo 10. Retroalimentación
La crítica deberá ser:

- Argumentada  
- Respetuosa  
- Metodológica  
- Constructiva  

No se permitirán **descalificaciones personales** ni juicios sin sustento académico.

---

## CAPÍTULO V  
### Conducta y sanciones

#### Artículo 11. Conducta esperada

- Respeto absoluto entre participantes.
- Lenguaje académico y profesional.
- Escucha activa y participación responsable.

#### Artículo 12. Faltas
Se consideran faltas graves:

- Plagio.
- Uso indebido de información.
- Violencia verbal o simbólica.
- Incumplimiento reiterado del reglamento.

#### Artículo 13. Sanciones
Las sanciones podrán incluir:

- Llamado de atención.
- Suspensión temporal.
- Expulsión definitiva sin reembolso.

---

## CAPÍTULO VI  
### Aceptación

#### Artículo 14. Aceptación del reglamento
La participación en la Clínica implica la **aceptación total del presente reglamento**.
